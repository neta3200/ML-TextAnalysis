```python
# YOUR CODE HERE
df_train['gender'] = df_train['gender'].replace(['f'], 0)
df_train['gender'] = df_train['gender'].replace(['m'], 1)
df_train.head(8)
```

```python
def get_amount_inst_class(df):
    cntf = df[df.gender == 0 ].shape[0]
    cntm = df[df.gender == 1].shape[0]
    
    print(f"overall train: {df.shape[0]}, male train: {cntm}, female train {cntf}")
    return cntm,cntf
```

```python
# Add as many code cells as you need
# YOUR CODE HERE

instances_f = df_train[df_train.gender == 0]

rows = instances_f.sample(n=int(cntm - cntf), replace=True, random_state=146)
df_train = df_train.append(rows)


cntm, cntf = get_amount_inst_class(df_train)
```

```python
X_series = pd.Series(df_train['story'])
y_series = pd.Series(df_train['gender'])

X_array = X_series.to_numpy()
y_array = y_series.to_numpy()


#print(X_array)
#print(y_array)
```

```python
def tok_func(hebrew_text):
    ret_tok = []
    tokens = ht.tokenize(hebrew_text)
    for grp, token, token_num, (start_index, end_index) in tokens:
        if grp == 'HEBREW':
            ret_tok.append(token)
    return ret_tok
```

```python
# Models imports for evaluation - all from sklearn

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import RidgeClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC

from numpy import mean
from numpy import isnan
```

```python
# A function to calculate our evalution method - "average_f1"
def calc_average_f1(y_true, y_pred):
    f1_male = metrics.f1_score(y_true, y_pred, pos_label=1)
    f1_female = metrics.f1_score(y_true, y_pred, pos_label=0)
    average_f1 = (f1_male + f1_female) / 2
    print(f"* Average_f1: {average_f1}\n* f1_male: {f1_male}\n* f1_female: {f1_female}")
    return average_f1
```

```python
# Getting a list of models to evaluate
def get_models():
    models = list()
    models.append(LogisticRegression())
    models.append(RidgeClassifier())
    models.append(SGDClassifier(shuffle=False, random_state=1))
    models.append(KNeighborsClassifier())
    models.append(DecisionTreeClassifier(random_state=1))
    models.append(LinearSVC())
    models.append(SVC())
    return models
```

```python
# Evaluating each model according to scores and mean score

def evaluate_model(cv, model):
    # evaluate the model
    text_clf = Pipeline([
        ('vect', TfidfVectorizer(tokenizer=tok_func)),
        ('norm', preprocessing.Normalizer(norm='l2')),
        ('clf', model),
    ])
    text_clf.fit(df_train.story, df_train.gender)
    scores = cross_val_score(text_clf, X=X_array, y=y_array, cv=cv, n_jobs=1, scoring='f1_macro')
    print(scores)
    # Return mean of scores
    return mean(scores)
```

```python
# Getting the list of models to consider
models = get_models()
#print(models)

# Collecting results
cv_results = list()

# Evaluating the model using a given test condition
kf = KFold(n_splits=10, shuffle=False, random_state=None)

# Evaluating each model
for model in models:
    # Evaluating model using each test condition
    cv_mean = evaluate_model(kf, model)
    # Checking invalid results
    if isnan(cv_mean):
        continue
    # Storing results
    cv_results.append(cv_mean)
    # Summarizing progress
    print('>%s: cv=%.3f' % (type(model).__name__, cv_mean))
```

```python
fold_train = []
fold_validation = []

for train_index, val_index in kf.split(X_array):
    X_train, X_val = X_array[train_index], X_array[val_index]
    y_train, y_val = y_array[train_index], y_array[val_index]
    
    # Re-arranging back into dataframes for train/validation
    df_train = pd.DataFrame({'story': X_train, 'gender':y_train})
    df_validation = pd.DataFrame({'story': X_val, 'gender':y_val})
    
    # Creating a pipeline with all of the steps above that were performed on the train set,
    # so we could easily use it for the validation & test datasets
    text_clf = Pipeline([
        ('vect', TfidfVectorizer(tokenizer=tok_func)),
        ('norm', preprocessing.Normalizer(norm='l2')),
        ('clf', SGDClassifier(shuffle=False, random_state=1)),
    ])
    # Training the model using the pipepline
    text_clf.fit(df_train.story, df_train.gender)
    
    # Testing on validation dataset
    predicted_validation = text_clf.predict(df_validation.story)
    
    # Calculating average_f1 for evaluation
    avg_f1 = calc_average_f1(df_validation.gender, predicted_validation)
    
    # Saving each fold and validation in the list 
    fold_train.append(df_train)
    fold_validation.append(df_validation)
```

```python
# Selecting fold & validation for best results

df_train = fold_train[1]
df_validation = fold_validation[1]
```

```python
# Training the model
text_clf.fit(df_train.story, df_train.gender)

# Testing on validation dataset
predicted_validation = text_clf.predict(df_validation.story)

# Printing evalution results on validation dataset
print("Evalution results on validation dataset:")
avg_f1 = calc_average_f1(df_validation.gender, predicted_validation)
```

```python
# Predicting test set
test_predicted = text_clf.predict(df_test.story)
df_predicted = df_test.copy()
df_predicted.loc[:,'predicted_category'] = pd.Series(test_predicted, index=df_predicted.index)
```
